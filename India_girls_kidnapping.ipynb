# 📌 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 🧹 2. Load the Dataset

```python
file_path = "India_girls_kidnapping_extended2.csv"  # Make sure the file is in your working directory
df = pd.read_csv(file_path)
```

# 🛠️ 3. Data Cleaning

```python
df['Girls Affected'] = pd.to_numeric(df['Girls Affected'], errors='coerce')
df = df.dropna(subset=['Girls Affected', 'State/Location'])
```

# 📊 4. Bar Plot: Girls Affected per Location

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='State/Location', y='Girls Affected', palette='magma')
plt.title("Girls Affected by Kidnapping per Location")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Number of Girls")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='State/Location', y='Girls Affected', palette='magma')
plt.title("Girls Affected by Kidnapping per Location")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Number of Girls")
plt.tight_layout()
plt.show()
```

# 🥧 5. Pie Chart: Top 6 Locations

```python
top6 = df.sort_values(by='Girls Affected', ascending=False).head(6)
plt.figure(figsize=(7, 7))
plt.pie(top6['Girls Affected'], labels=top6['State/Location'], autopct='%1.1f%%', startangle=140)
plt.title("Top 6 Locations by % of Girls Affected")
plt.axis('equal')
plt.show()
```

# 📈 6. Trend Line (if year data is present)

```python
df['Year_Extracted'] = df['Year'].str.extract(r'(\d{4})').astype('float')
yearly = df.dropna(subset=['Year_Extracted']).groupby('Year_Extracted')['Girls Affected'].sum().reset_index()
```

```python
plt.figure(figsize=(10, 5))
sns.lineplot(data=yearly, x='Year_Extracted', y='Girls Affected', marker='o')
plt.title('Yearly Trend of Girls Affected by Kidnapping')
plt.xlabel('Year')
plt.ylabel('Total Girls Affected')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 🔍 7. Show Summary Table

```python
summary = df.groupby("State/Location")['Girls Affected'].sum().sort_values(ascending=False).reset_index()
print("📌 Summary of Girls Affected by Location:")
print(summary)
```

# 1. Donut Chart (Modified Pie Chart)


```python
top_states = df.groupby('State/Location')['Girls Affected'].sum().sort_values(ascending=False).head(5)
plt.figure(figsize=(6, 6))
plt.pie(top_states, labels=top_states.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
plt.title("Top 5 States - Girls Affected (Donut Chart)")
plt.axis('equal')
plt.show()
```

# Violin Plot by State (To Show Distribution)

```python
plt.figure(figsize=(12, 6))
sns.violinplot(data=df, x='State/Location', y='Girls Affected', palette='coolwarm')
plt.xticks(rotation=45, ha='right')
plt.title("Distribution of Girls Affected by State (Violin Plot)")
plt.tight_layout()
plt.show()
```
